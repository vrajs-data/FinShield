```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import mysql.connector
```

```python
np.random.seed(42)
```

```python
rows=10000
```

```python
data = []
```

```python
locations = ['India', 'USA', 'UK', 'Canada', 'Germany']
devices = ['Mobile', 'Desktop', 'Tablet']
categories = ['Food', 'Shopping', 'Bill', 'Travel', 'Entertainment']
payments = ['Card', 'UPI', 'Netbanking', 'Wallet']
```

```python
for i in range(rows):
    user_id = np.random.randint(1000,2000)
    amount = round(np.random.exponential(200), 2)
    time = datetime.now() - timedelta(minutes = np.random.randint(0, 100000))
    is_int = np.random.choice([0,1], p=[0.8,0.2])
    gap = np.random.randint(1,60)
    avg_spend = round(np.random.uniform(100, 500),2)
    freq = np.random.randint(1,20)

    is_fraud = 1 if (amount > avg_spend*2 and gap < 5 and is_int == 1) else 0

    data.append([
        i, user_id, amount, time,
        random.choice(categories),
        random.choice(locations),
        random.choice(devices),
        random.choice(payments),
        is_int, gap, avg_spend,freq, is_fraud
    ])

df = pd.DataFrame(data , columns=[
    'transaction_id', 'user_id', 'transaction_amount', 'transaction_time', 'merchant_catrgory', 'location', 'device_type', 'payment_method',
    'is_international', 'previous_transaction_gap', 'user_avg_spend', 'transaction_frequency', 'is_fraud'
])

df.to_csv("transactions.csv", index=False)
```
